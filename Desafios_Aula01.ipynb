```
#carregando dados e importando biblioteca
import pandas as pd

fonte = "https://github.com/nickvgs/data_imersao_dados_alura_2ed/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true"

dados = pd.read_csv(fonte)
dados.head(80)
```

Desafio01: Proporção dos inscritos por idade.



```
#vendo a proporção de inscritos por idade ordenado por index
dados["NU_IDADE"].value_counts(normalize=True).sort_index()
```

Desafio02: Descobrir de quais estados são os inscritos com 13 anos.


```
dados.query("NU_IDADE == 13")["SG_UF_RESIDENCIA"].value_counts()
```

Desafio03: Adicionar título no gráfico




```
#histograma das notas de redação e definição de titulo
dados['NU_NOTA_REDACAO'].hist(bins=20, figsize=(12,6)).set_title('Histograma das Notas de Redação', fontsize=20)
```

Desafio04: Plotar os Histogramas das idades dos do treineiro e não treineiros.


```
#CROSSTAB TREINEIRO/IDADE
pd.crosstab(index=dados['NU_IDADE'], columns=dados['IN_TREINEIRO']).hist();
```
